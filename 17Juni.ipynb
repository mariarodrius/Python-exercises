```python
df = pd.read_csv('https://query.data.world/s/5dpnyov52z2lsghtptp4v6n473tm6l')
newdf=pd.melt(df, id_vars = 'Item',var_name='Year', value_name ='No of Transactions')
newdf
```

```python
print(newdf.info())
```

```python
print(df.info())  #265 Items shown per year
```

```python
#https://thispointer.com/python-pandas-how-to-drop-rows-in-dataframe-by-conditions-on-column-values/
indexNames = newdf[newdf['Item'] == "TOTAL"].index 
newdf.drop(indexNames , inplace=True)
print(newdf.info())
```

```python
newdf["No of Transactions"]=newdf["No of Transactions"]/1000000
newdf=newdf.round(1)
newdf
```

```python
f, ax = plt.subplots(figsize = (20,80))
sns.set_color_codes('pastel')
sns.barplot(x = 'No of Transactions', y = 'Item', data = newdf,
            label = 'No of Transactions', color = 'g', edgecolor = 'w')
sns.set_color_codes('muted')
sns.barplot(x = 'Year', y = 'Item', data = newdf,
            label = 'No of Transactions', color = 'b', edgecolor = 'w')
plt.show()
```
